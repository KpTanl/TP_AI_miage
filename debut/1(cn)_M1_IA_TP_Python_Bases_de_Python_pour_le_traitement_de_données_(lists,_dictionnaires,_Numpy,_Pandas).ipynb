## Indexing
## 索引





Une **liste** est une collection ordonnée de différents objets tels que des entiers, des chaînes de caractères et même d’autres listes.
**列表**是不同对象（如整数、字符串甚至其他列表）的有序集合。

L’adresse de chaque élément dans une liste est appelée un <b>index</b>. Un index est utilisé pour accéder aux éléments d’une liste et s’y référer.
列表中每个元素的地址称为<b>索引</b>。索引用于访问和引用列表中的元素。

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/ListsIndex.png" width="1000">


Pour créer une liste, écrivez la liste entre crochets <b>[ ]</b>, avec son contenu à l’intérieur des parenthèses et séparé par des virgules.
要创建列表，请在方括号 <b>[ ]</b> 中写入列表，内容在括号内并用逗号分隔。

````python
# Créer une liste / 创建列表

L = ["The Bodyguard", 7.0, 1992]
L
````

On peut ainsi accéder à un élément par son index (**indexing**).
我们可以通过索引访问元素（**索引**）。

````python
L = ["The Bodyguard", 7.0, 1992]
L[2]
````

A noter que l'index peut être **négatif**. L'index -1 pointe vers le dernier élément de la liste, -2 vers l'avant-dernier...
请注意，索引可以是**负数**。索引 -1 指向列表的最后一个元素，-2 指向倒数第二个...

````python
L[-1]
````

## Opérations sur les listes
## 列表操作


Nous pouvons effectuer des opérations de découpage (**slicing**) sur les listes. Par exemple, si nous voulons les deux derniers éléments, nous utilisons la commande suivante :
我们可以对列表执行切片操作 (**slicing**)。例如，如果我们想要最后两个元素，我们使用以下命令：





````python
# Exemple de liste

L = ["The Bodyguard", 7.0,1992,"BG",1]
L
````

````python
# List slicing / 列表切片

L[3:5]
````

Il est également possible de ne rien préciser à gauche ou à droite du symbole `:`:
如果在符号 `:` 的左侧或右侧未指定任何内容，也是可以的：
* `:2` retourne tous les éléments du début de la liste jusqu'à 2 (exclus)
* `:2` 返回从列表开头到 2（不包括）的所有元素
* `2:` retourne tous les éléments depuis l'indice 2 jusqu'au bout de la liste.
* `2:` 返回从索引 2 到列表末尾的所有元素。

````python
L[:2]
````

````python
L[2:]
````

A noter que le slicing permet également de **copier** une liste.
请注意，切片也可以用来**复制**列表。

Une copie d'une liste peut être modifiée sans que la liste d'origine ne le soit.
可以在不修改原始列表的情况下修改列表的副本。

````python
copie_L = L[:]
copie_L[0] = "Hitman"
print(copie_L)
print(L)
````

Comme les listes sont **mutables**, nous pouvons les modifier.
由于列表是**可变**的，我们可以修改它们。
Note: ceci n'est pas le cas des tuples, qui sont **immutables**.
注意：这不适用于**不可变**的元组。
Par exemple, nous pouvons changer le premier élément comme suit :
例如，我们可以按如下方式更改第一个元素：

````python
# Change the element based on the index / 根据索引更改元素

A = ["disco", 10, 1.2]
print('Avant modification:', A)
A[0] = 'hard rock'
print('Après modification:', A)
````

Nous pouvons utiliser la méthode `append` pour ajouter un nouvel élément à la liste.
我们可以使用 `append` 方法向列表添加新元素。

````python
L = [ "The Bodyguard", 7.0]
L.append('pop')
print(L)
L.append('10')
print(L)
L.append(['carotte', 20])
print(L)
````

La méthode `append` ajoute chaque élément passé en paramètre à la liste, même si l'élément est une liste.
 `append` 方法将作为参数传递的每个元素添加到列表中，即使该元素是列表。
Nous pouvons utiliser la méthode `extend` pour **concaténer** les listes:
我们可以使用 `extend` 方法来**连接**列表：

````python
# Concaténer 2 listes / 连接 2 个列表

L = [ "The Bodyguard", 7.0]
L.extend(['pop', 10])
L
````

Nous pouvons aussi supprimer un élément de la liste en utilisant la fonction `del`:
我们也可以使用 `del` 函数从列表中删除元素：

````python
# Supprimer un élément à partir de son index / 根据索引删除元素

A = ["disco", 10, 1.2]
print('Avant modification:', A)
del(A[0])
print('Après modification:', A)
````

Nous pouvons convertir une chaîne de caractères en liste en utilisant `split`. Par exemple, la méthode `split` transforme chaque groupe de caractères séparé par un **delimiter** choisi en un élément de la liste :
我们可以使用 `split` 将字符串转换为列表。例如，`split` 方法将由所选**分隔符**分隔的每组字符转换为列表元素：

````python
# Split une chaîne de caractère selon le caractère espace (delimiter) / 根据空格字符（分隔符）拆分字符串

'hard rock'.split(" ")
````

## Qu'est-ce qu'un dictionnaire ?


Un dictionnaire est une structure de données qui se compose de **clés** et de **valeurs**. Ces clés permettent d’accéder aux valeurs contenues dans le dictionnaire.
字典是由**键**和**值**组成的数据结构。这些键允许访问字典中包含的值。

Dans une liste, les éléments sont indexés par un nombre entier, alors que dans un dictionnaire l'indexe est une clé, qui peut être de n'importe quel type (où plus précisément de tout type immutable, comme un entier, une string, un tuple ...).
在列表中，元素由整数索引，而在字典中，索引是一个键，可以是任何类型（或者更准确地说是任何不可变类型，如整数、字符串、元组...）。

Le meilleur exemple d’un dictionnaire est l’accès aux informations d’une personne à l’aide du **numéro de sécurité sociale**.
字典最好的例子是使用**社会安全号码**访问有关一个人的信息。
Ici, le numéro de sécurité sociale, qui est unique, sera la **clé**, et les informations de la personne seront les **valeurs** qui y sont associées.

## Créer un Dictionaire et accéder aux éléments
## 创建字典并访问元素

Ici, nous créons un dictionnaire nommé **Dict** avec les détails suivants :
在这里，我们创建一个名为 **Dict** 的字典，详细信息如下：

* Les clés sont **"key1", "key2", "key3", "key4", "key5" et (0,1)**.
* 键是 **"key1", "key2", "key3", "key4", "key5" 和 (0,1)**。
* Les valeurs sont 1, 2, [3, 3, 3], (4, 4, 4), 5 et 6.


````python
# Créer un dictionaire / 创建字典

Dict = {"key1": 1, "key2": "2", "key3": [3, 3, 3], "key4": (4, 4, 4), 'key5': 5, (0, 1): 6}
Dict
````

Les clés peuvent être des chaînes de caractères :


````python
# Accéder à une valeur par sa clé

Dict["key1"]
````

Les clés peuvent aussi être de tout type immutable, comme un tuple :
键也可以是任何不可变类型，例如元组：


````python
# Accéder à une valeur par sa clé
# 通过键访问值

Dict[(0, 1)]
````

Chaque clé est séparée de sa valeur par deux-points "<code>:</code>". Les éléments sont séparés par des virgules, et l’ensemble du dictionnaire est encadré par des accolades. Un dictionnaire vide, sans aucun élément, s’écrit simplement avec deux accolades, comme ceci : "<code>{}</code>".
每个键与其值之间用冒号“<code>:</code>”分隔。元素之间用逗号分隔，整个字典用大括号括起来。一个没有任何元素的空字典只需用两个大括号书写，如下所示：“<code>{}</code>”。


````python
# Créer / 创建

release_year_dict = {"Thriller": "1982", "Back in Black": "1980", \
                    "The Dark Side of the Moon": "1973", "The Bodyguard": "1992", \
                    "Bat Out of Hell": "1977", "Their Greatest Hits (1971-1975)": "1976", \
                    "Saturday Night Fever": "1977", "Rumours": "1977"}
release_year_dict
````

En résumé, comme une liste, un dictionnaire contient une séquence d’éléments. Chaque élément est représenté par une **clé** et sa **valeur** correspondante. **Pour chaque clé, il ne peut y avoir qu’une seule valeur**, mais plusieurs clés peuvent partager la même valeur. Les clés ne peuvent être que des chaînes de caractères, des nombres ou des tuples, tandis que les valeurs peuvent être de n’importe quel type de données.
总之，就像列表一样，字典包含一系列元素。每个元素由一个**键**及其对应的**值**表示。**对于每个键，只能有一个值**，但多个键可以共享相同的值。键只能是字符串、数字或元组，而值可以是任何数据类型。


Il est utile de visualiser le dictionnaire comme un tableau, comme dans l’image suivante. La première colonne représente les **clés**, et la deuxième colonne représente les **valeurs**.
将字典可视化为表格很有用，如下图所示。第一列代表**键**，第二列代表**值**。

## Clés
## 键


On peut accéder aux valeurs d'un dictionnaire par leur clé :
我们可以通过键访问字典的值：

````python
# Récupérer une valeur par sa clé / 通过键获取值

release_year_dict['Thriller']
````

Cela correspond à :


<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%202/images/DictsKeyOne.png" width="500">


On peut **modifier la valeur** correspondant à une clé :
我们可以**修改值**对应的键：

````python
# Modifier la valeur correspondant à une clé / 修改键对应的值
release_year_dict['Thriller'] = "1983"
release_year_dict
````

Dans le cas où la clé n'existe pas dans l'exemple précédent, la **paire clé valeur est ajoutée** au dictionnaire :
如果前面的示例中不存在该键，则将**键值对添加到**字典中：

````python
# Ajouter une paire clé/valeur au dictionnaire ('Graduation' n'est pas une clé du dictionnaire) / 向字典添加键/值对（'Graduation' 不是字典的键）

release_year_dict['Graduation'] = '2007'
release_year_dict
````

On peut **supprimer** un élément du dictionnaire avec la fonction `del` :


````python
# Delete entries by key / 按键删除条目

del(release_year_dict['Thriller'])
del(release_year_dict['Graduation'])
release_year_dict
````

Enfin, on peut vérifier qu'un élément une clé du dictionnaire avec l'opérateur `in` :
最后，我们可以使用 `in` 运算符检查元素是否是字典的键：

````python
# Vérifie si la clé est dans le dictionnaire / 检查键是否在字典中

'The Bodyguard' in release_year_dict
````

<a id="dict-boucle"></a>
## Dictionnaire et boucles


On peut récupérer toutes les clés d'un dictionnaire avec la méthode <code>keys()</code>:


````python
# Récupérer toutes les clés / 获取所有键

release_year_dict.keys()
````

Utilisée dans une boucle, cette méthode nous permet de parcourir tous les éléments d'un dictionnaire :
在循环中使用此方法，我们可以遍历字典的所有元素：

````python
# Afficher toutes les paires d'un dictionnaire / 显示字典的所有对
for k in release_year_dict.keys():
  print(k,"-",release_year_dict[k])
````

On peut également récupérer toutes les valeurs d'un dictionnaire avec la méthode `values()`.
我们也可以使用 `values()` 方法检索字典的所有值。

A noter qu'il n'est pas possible de retrouver la clé à partir d'une valeur, car plusieurs clés différentes peuvent correspondre à la même valeur.


````python
# Get all the values in dictionary / 获取字典中的所有值

release_year_dict.values()
````

Enfin on peut récupérer toutes les paires avec la méthode `items()`:
最后，我们可以使用 `items()` 方法检索所有对：

````python
release_year_dict.items()
````

````python
for k,v in release_year_dict.items():
  print(k,"-",v)
````

## Qu'est-ce que Numpy?


NumPy est une bibliothèque Python utilisée pour travailler avec des **tableaux** (à 1 ou n dimensions) et de réaliser des opérations d’**algèbre linéaire**... NumPy signifie *Numerical Python* et c’est un projet open source. L’objet tableau dans NumPy s’appelle **ndarray** et il fournit de nombreuses fonctions qui facilitent grandement le travail avec les ndarrays.
NumPy 是一个 Python 库，用于处理**数组**（1 维或 n 维）并执行**线性代数**操作... NumPy 意为 *Numerical Python*，是一个开源项目。NumPy 中的数组对象称为 **ndarray**，它提供了许多函数，可以极大地方便 ndarray 的处理。

Les tableaux sont très fréquemment utilisés en **science des données**, où la vitesse et les ressources sont très importantes.
数组在**数据科学**中非常常用，其中速度和资源非常重要。

### Slicing


Comme pour les listes, nous pouvons effectuer des opérations de découpage (*slicing*) sur les tableaux NumPy. Le découpage en Python consiste à extraire des éléments compris entre un index de départ et un index de fin donné.
与列表一样，我们可以对 NumPy 数组执行切片 (*slicing*) 操作。Python 中的切片包括提取给定开始索引和结束索引之间的元素。

Nous passons le découpage de la manière suivante : `[start:end]`. L’élément à l’index de fin n’est **pas inclus** dans le résultat.
我们按如下方式传递切片：`[start:end]`。结果中**不包含**结束索引处的元素。

Nous pouvons sélectionner les éléments de 1 à 3 et les affecter à un nouveau tableau NumPy <code>d</code> comme suit :


````python
# Slicing le tableau Numpy / 对 Numpy 数组进行切片

d = c[1:4]
d
````

On peut assigner des nouvelles valeur aux indexes précisés dans le slicing :
我们可以为切片中指定的索引分配新值：

````python
# Assigne au 4eme et 5eme éléments les valeurs 300 et 400 / 为第 4 和第 5 个元素分配值 300 和 400

c[3:5] = 300, 400
c
````

On peut aussi définir un pas dans le découpage avec : `[start:end:step]`.


````python
arr = np.array([1, 2, 3, 4, 5, 6, 7])

print(arr[1:5:2])
````

Si *start* n'est pas précisé, sa valeur par défaut est 0.


````python
print(arr[:4])
````

Si *end* n'est pas précisé, sa valeur par défaut est la taille du tableau.


````python
print(arr[4:])
````

Si *step* n'est pas précisé, sa valeur par défaut est 1.


````python
print(arr[1:5:])
````

### Fonctions Numpy
### Numpy 函数

Numpy propose également des fonctions permettant de réalisant les différentes opérations arithmétiques de base :
Numpy 还提供了允许执行各种基本算术运算的函数：
* `np.add(arr1,arr2)`
* `np.subtract(arr1, arr2)`
* `np.multiply(arr1, arr2)`
* `np.divide(arr1, arr2)`



````python
arr1 = np.array([10, 11, 12, 13, 14, 15])
arr2 = np.array([20, 21, 22, 23, 24, 25])

addi = np.add(arr1,arr2)
print(addi)
sub = np.subtract(arr1, arr2)
print(sub)
mult = np.multiply(arr1, arr2)
print(mult)
div = np.divide(arr1, arr2)
print(div)
````

Les fonctions Numpy (ex. `np.add()`) sont équivalentes aux opérateurs de bases (ex. `+`) dans beaucoup de cas.
在许多情况下，Numpy 函数（例如 `np.add()`）等同于基本运算符（例如 `+`）。

Néanmoins les fonctions Numpy permettent des options supplémentaires comme:
但是，Numpy 函数允许额外的选项，例如：
* `where` permet de n'effectuer l'opération que lorsque la condition est vérifiée. Pour les autres éléments, la valeur n'est pas définie
* `where` 允许仅在满足条件时执行操作。对于其他元素，未定义该值
* `out` spécifie dans quel tableau le résultat est placé, sans avoir à le créer.
* `out` 指定将结果放置在哪个数组中，而无需创建它。

Ces options permettent d'optimiser des gros calculs.
这些选项允许优化大型计算。

````python
arr1 = np.array([10, 11, 12, 13, 14, 15])
arr2 = np.array([20, 21, 22, 23, 24, 25])

r = np.zeros_like(arr1)
np.add(arr1,arr2,out=r,where=arr1 > 13)
````

La multiplication de deux tableaux correspond à un produit élément par élément ou <em>produit de Hadamard</em>.
两个数组的乘法对应于逐元素乘积或 <em>Hadamard 积</em>。
Considérons les matrices <code>X</code> et <code>Y</code>. Le produit de Hadamard consiste à multiplier **chaque élément situé à la même position**, c’est-à-dire à multiplier ensemble les éléments contenus dans les mêmes cases colorées.
考虑矩阵 <code>X</code> 和 <code>Y</code>。Hadamard 积包括乘以**位于相同位置的每个元素**，即将在相同颜色方框中包含的元素相乘。
Le résultat est une nouvelle matrice de **la même taille** que la matrice <code>Y</code> ou <code>X</code>, comme illustré dans la figure suivante.

````python
# Import required library

import pandas as pd
````

**Pandas** est une librairie qui utilise extensivement Numpy. Elle permet de manipuler des données tabulaires sous la forme d'un **dataframe**. Cette structure de données peut être décrite comme une matrice avec des noms de colonnes et de lignes.
**Pandas** 是一个广泛使用 Numpy 的库。它允许以 **dataframe** 的形式操作表格数据。这种数据结构可以描述为具有列名和行名的矩阵。

Après la commande d’importation, nous avons désormais accès à un grand nombre de **classes et fonctions préconstruites**. Cela suppose que la bibliothèque est installée ; dans notre environnement de laboratoire, toutes les bibliothèques nécessaires sont déjà installées.

Une façon dont Pandas permet de travailler avec les données est via un **dataframe**. Voyons le processus pour passer d’un fichier de valeurs séparées par des virgules (<b>.csv</b>) à un dataframe.
Pandas 允许处理数据的一种方式是通过 **dataframe**。让我们看看从逗号分隔值文件 (<b>.csv</b>) 到 dataframe 的过程。

La variable <code>csv_path</code> stocke le chemin du fichier <b>.csv</b>, qui est utilisé comme argument pour la fonction <code>read_csv</code>. Le résultat est stocké dans l’objet <code>df</code>, forme courte couramment utilisée pour une variable faisant référence à un dataframe Pandas.
