```python
#python obj


class ReseauSocial:
    def __init__(self):
        self.reseau = {}
        # {"name":"friends"}
    def adduser(self,user):
        if user not in self.reseau:
            self.reseau[user] = []
    def __str__(self):
        res = "Reseau social :"
        for user,friends in self.reseau.items():
            res += f"\n{user} -> {','.join(friends)} "
        return res
    def addlink(self,user1,user2):
        if user1 not in self.reseau:
            self.adduser(user1)
        if user2 not in self.reseau:
            self.adduser(user2)
        if user2 not in self.reseau[user1]:
            self.reseau[user1].append(user2)
        if user1 not in self.reseau[user2]:
            self.reseau[user2].append(user1)
    def get_users(self):
        return list(self.reseau.keys())
    # B1    
    def is_friends(self,user1,user2):
        if user1 in self.reseau:
            if user2 in self.reseau[user1]:
                return True
        return False
    # B2    
    def most_friends(self):
        max_count = -1
        popular_user = None
        for user, friends in self.reseau.items():
            count = len(friends)
            if count > max_count:
                max_count = count
                popular_user = user
        return popular_user
    # B3
    def same_friends(self, user1, user2):
        common_list = []
        friends1 = self.reseau.get(user1, [])
        friends2 = self.reseau.get(user2, [])
        for friend in friends1:
            if friend in friends2 and friend not in common_list:
                common_list.append(friend)
        return common_list
    
    # B4
    def recommend_friends(self,user):
        isnt_friend_list = []
        for us in self.reseau.keys():
            if us != user and us not in self.reseau[user]:
                isnt_friend_list.append(us)
        max_same_friend_count = -1
        reco_friend = None
        for us in isnt_friend_list:
            if len(self.same_friends(us,user)) > max_same_friend_count:
                max_same_friend_count = len(self.same_friends(us,user)) 
                reco_friend = us
        return reco_friend
    
    #B5 
    # def find_distance BFS：
    def distance(self,user1,user2):
        if user1 not in self.reseau or user2 not in self.reseau:
            return -1
        elif user1 == user2:
            return 0
        else:
            #debut
            queue = [(user1,0)] 
            visited = {user1}
            curr , dis = queue.pop(0)
            #while
            while queue:
                # pochur chaque neoud
                for fr in self.reseau[curr]:
                    if fr == user2 :
                        return  dis + 1
                    if fr not in visited:
                        visited.append(fr)
                        queue.append(fr,dis + 1)
            return -1
    
    # DFS?
    def distance_dfs(self,user1,user2):
        if user1 not in self.reseau or user2 not in self.reseau:
            return -1
        if user1 == user2:
            return 0
        
        def dfs(curr,tar,visited):
            if curr == tar :
                return 0
            visited.append(curr)
            best = None
            for fr in self.reseau[curr]:
                if fr not in visited:
                    d = dfs (fr, tar, visited)
                if d is not None or d + 1 < best:
                    best = d + 1
                visited.remove(curr) #
            return best
        ans =  dfs (user1,user2,set())
        return -1 if ans is None else ans




        
```

```python
# main
ts = ReseauSocial()
ts.adduser("A")
ts.adduser("B")
ts.adduser("C")
ts.addlink("A","B")
ts.addlink("A","C")
print(ts)
print(ts.is_friends("A","C"))
print(ts.is_friends("B","C"))
print(ts.same_friends("A","C"))
print(ts.same_friends("B","C"))
```
